```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-colorblind")
```

```python
vals = [0.3, 0.3, 0.4]
entropy = - np.sum(vals * np.log(vals))
print(entropy)

vals = [0.9, 0.05, 0.05]
entropy = - np.sum(vals * np.log(vals))
print(entropy)

vals = [0.99, 0.005, 0.005]
entropy = - np.sum(vals * np.log(vals))
print(entropy)
```

```python
vals = [0.3, 0.3, 0.4]

x = np.linspace(0, 1, 100)
y = np.empty(100, dtype=np.float32)
for i, p in enumerate(x):
    y[i] = np.sum(np.abs(vals) ** p)
    
plt.plot(x, y)
```
